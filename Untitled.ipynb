##  Replacing nan of dependant column and presenting problematic rows

```python

df = pd.DataFrame(dataset)

# Get the indices of NaN values in 'dependants' inform of a list.
nan_indices = df[df['Dependants'].isna()].index.tolist()  

# Calculate the mean of the column
mean_value = df['Dependants'].mean()

# Replace NaN values in 'Dependants' with the mean
df['Dependants'].fillna(mean_value, inplace=True)

# Print the DataFrame after replacing NaN values
print("dataset:")
print(df)

# Print the indices of replaced NaN values
print("Indices of replaced NaN values:", nan_indices)

#print(nan_indices)
```
